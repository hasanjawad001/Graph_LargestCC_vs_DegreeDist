```python
import networkx as nx
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
def draw_size_vs_deg(n=100):
    list_p = [i/(20*n) for i in range(10,61,1)]
    list_x = []
    for p in list_p:
        list_x.append(p*(n-1))
    # print(list_p)
    # print(list_x)
    list_y = []
    for p in list_p:
        G = nx.gnp_random_graph(n, p, seed=123, directed=False)
        # nx.draw(G)
        # identify largest connected component
        Gcc = sorted(nx.connected_components(G), key=len, reverse=True)
        G0 = G.subgraph(Gcc[0])
        n0 = G0.number_of_nodes()
        frac0 = n0/n
        list_y.append(frac0)

    fig1 = plt.figure(constrained_layout=True)
    spec1 = gridspec.GridSpec(ncols=1, nrows=1, figure=fig1)
    ax1 = fig1.add_subplot(spec1[0, 0])
    exp_title = 'Exp (n=' + str(n) + ') : fraction of nodes in largest CC vs [p*(n-1)]'
    fig1.suptitle(exp_title)
    ax1.plot(list_x, list_y, 'o-')
    ax1.set_xlabel('[p*(n-1)]')
    ax1.set_ylabel('fraction of nodes in largest CC')
    # print()
    #############################################################################################
def draw_size_vs_deg_large_interval(n=1000):
    list_p = [i/(20*n) for i in range(10,61,10)]
    list_x = []
    for p in list_p:
        list_x.append(p*(n-1))
    # print(list_p)
    # print(list_x)
    list_y = []
    for p in list_p:
        G = nx.gnp_random_graph(n, p, seed=123, directed=False)
        # nx.draw(G)
        # identify largest connected component
        Gcc = sorted(nx.connected_components(G), key=len, reverse=True)
        G0 = G.subgraph(Gcc[0])
        n0 = G0.number_of_nodes()
        frac0 = n0/n
        list_y.append(frac0)

    fig1 = plt.figure(constrained_layout=True)
    spec1 = gridspec.GridSpec(ncols=1, nrows=1, figure=fig1)
    ax1 = fig1.add_subplot(spec1[0, 0])
    exp_title = 'Exp (n=' + str(n) + ') : fraction of nodes in largest CC vs [p*(n-1)]'
    fig1.suptitle(exp_title)
    ax1.plot(list_x, list_y, 'o-')
    ax1.set_xlabel('[p*(n-1)]')
    ax1.set_ylabel('fraction of nodes in largest CC')
    # print()
    #############################################################################################
```
